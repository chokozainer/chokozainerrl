#学習
DQN (Deep Q Learning)
 
右のテーブルを変更すると、左に反映されます。

```
#@title Configuration
args=chokozainerrl.train_a2c_gym.make_args([])

args.mode = 'train' #@param ["check", "train"]
args.outdir=thisdir #@param{type:"string"}
args.env = "CartPole-v0" #@param ["Acrobot-v1","Pendulum-v0","CartPole-v0","MountainCar-v0"]{allow-input: true}
args.gpu=-1 #@param{type:"integer"}
args.load_agent=None #@param{type:"string"}
args.steps=100000 #@param{type:"integer"}
args.step_offset=0 #@param{type:"integer"}
args.checkpoint_freq=10000 #@param{type:"integer"}
args.eval_interval=10000 #@param{type:"integer"}
args.eval_n_runs=10 #@param{type:"integer"}
args.reward_scale_factor=0.01 #@param{type:"number"}
args.arch='FFSoftmax' #@param['Gaussian','FFSoftmax','FFMellowmax']
args.seed=0 #@param{type:"integer"}
args.profile=False #@param{type:"boolean"}
args.update_steps=5 #@param{type:"integer"}
args.long_interval=1000 #@param{type:"integer"}
args.rmsprop_epsilon=0.00001 #@param{type:"number"}
args.render=False #@param{type:"boolean"}
args.gamma=0.99 #@param{type:"number"}
args.use_gae=False #@param{type:"boolean"}
args.tau=0.95 #@param{type:"number"}
args.lr=0.0007 #@param{type:"number"}
args.weight_decay=0.0 #@param{type:"number"}
args.max_grad_norm=0.5 #@param{type:"number"}
args.alpha=0.99 #@param{type:"number"}
args.num_envs=8 #@param{type:"integer"}
args.log_type="pbar"#@param["pbar","full_stream"]{allow-input: true}

```

##Train
訓練開始です。

```
chokozainerrl.train_a2c_gym.main(args)
```

##Result
学習結果の表です。

```
chokozainerrl.show_result.table(args.outdir+'/scores.txt')
```

学習結果のグラフです。

```
chokozainerrl.show_result.graph(args.outdir+'/scores.txt','steps','average_entropy')
```
