#前処理

```
import chokozainerrl
from pyvirtualdisplay import Display
display = Display(visible=0, size=(1024, 768))
display.start()
%matplotlib inline
```

#設定
右のテーブルを変更すると、左に反映されます。

```
#@title Configuration
args=chokozainerrl.train_pcl_gym.make_args([])

args.mode = 'train' #@param ["check", "train"]
args.processes=1 #@param{type:"integer"}
args.outdir='mydrive/OpenAI/CartPole/result_pcl' #@param{type:"string"}
args.env = "Pendulum-v0" #@param ["Pendulum-v0","CartPole-v0"]{allow-input: true}
args.gpu=-1 #@param{type:"integer"}
args.load_agent=None #@param{type:"string"}
args.save_agent='mydrive/OpenAI/CartPole/result_pcl/agent' #@param{type:"string"}
args.steps=80000000 #@param{type:"integer"}
args.step_offset=0 #@param{type:"integer"}
args.checkpoint_freq=10000 #@param{type:"integer"}
args.seed=0 #@param{type:"integer"}
args.batchsize=10 #@param{type:"integer"}
args.rollout_len=10 #@param{type:"integer"}
args.n_hidden_channels=100 #@param{type:"integer"}
args.n_hidden_layers=2 #@param{type:"integer"}
args.n_times_replay=1 #@param{type:"integer"}
args.t_max=None #@param{type:"integer"}
args.tau=0.02 #@param{type:"number"}
args.profile=False #@param{type:"boolean"}
args.eval_n_runs=10 #@param{type:"integer"}
args.eval_interval=100000 #@param{type:"integer"}
args.reward_scale_factor=0.01 #@param{type:"number"}
args.render=False #@param{type:"boolean"}
args.lr=0.0007 #@param{type:"number"}
args.monitor=False #@param{type:"boolean"}
args.train_async=False #@param{type:"boolean"}
args.prioritized_replay=False #@param{type:"boolean"}
args.disable_online_update=False #@param{type:"boolean"}
args.backprop_future_values=False #@param{type:"boolean"}
```

#Train
訓練開始です。

```
chokozainerrl.train_pcl_gym.main(args)
```

#Result
学習結果の表です。

```
chokozainerrl.show_result.table(args.outdir+'/scores.txt')
```

学習結果のグラフです。

```
chokozainerrl.show_result.graph(args.outdir+'/scores.txt','steps','average_q')
```
